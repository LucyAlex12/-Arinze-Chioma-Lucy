**Section 3: Feature Scaling**

Feature scaling is the process by which the value range of the data is adjusted to make sure that all features lie within the same range. The reason why this needs to be done is that machine learning algorithms can be biased towards the magnitude of the features.

Without feature scaling, a feature such as area or price, which has a large magnitude, will have a greater influence than another feature such as number of bedrooms, which has a lower magnitude.

**Question 10**

```python
def min_max_scale(col):
    return (col - col.min()) / (col.max() - col.min())

df['area_scaled'] = min_max_scale(df['area'])
df['bedrooms_scaled'] = min_max_scale(df['bedrooms'])
df['bathrooms_scaled'] = min_max_scale(df['bathrooms'])
df['price_scaled'] = min_max_scale(df['price'])
```

**Question 11**

```python
for col in ['area_scaled', 'bedrooms_scaled', 'bathrooms_scaled', 'price_scaled']:
    print(col, df[col].min(), df[col].max())
```
